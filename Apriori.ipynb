```python
#Importing necessary libraries
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
#Importing the dataset
df = pd.read_csv("/content/drive/MyDrive/4th SEM/ML/Datasets/Market_Basket_Optimisation - Market_Basket_Optimisation.csv", header=None)
df.head()
```

```python
df.shape
```

```python
#Filling missing values
df.fillna(0, inplace=True)
```

```python
transactions = []
for i in range(0,len(df)):
  transactions.append([str(df.values[i,j]) for j in range(0,20) if str(df.values[i,j])!="0"])
```

```python
transactions[0]
```

```python
#Implementing Apriori
from apyori import apriori
rules = apriori(transactions, min_support=0.003, min_confidance=0.2, min_lift=3, min_length=2)
Results = list(rules)
df_results = pd.DataFrame(Results)
df_results.head()
```

```python
#Extracting values from the dataset “df_results”
support = df_results.support
first_values = []
second_values = []
third_values = []
fourth_value = []
for i in range(df_results.shape[0]):
   single_list = df_results['ordered_statistics'][i][0]
   first_values.append(list(single_list[0]))
   second_values.append(list(single_list[1]))
   third_values.append(single_list[2])
   fourth_value.append(single_list[3])
```

```python
 # Processing the extracted values from the association rules into separate DataFrames
 lhs = pd.DataFrame(first_values)
 rhs= pd.DataFrame(second_values)
 confidance=pd. DataFrame(third_values , columns=[ 'Confidance'])
 lift = pd.DataFrame(fourth_value,columns = ['lift'])
 df_final = pd.concat([lhs, rhs, support, confidance, lift], axis = 1)
 df_final
```

```python
 df_final.fillna(value = " ",inplace = True)
 df_final.columns = ['lhs','lhs2','rhs','rhs2','rhs3','support','confidance','lift']
 df_final['lhs'] = df_final['lhs']+str(", ")+df_final['lhs2']
 df_final['rhs'] = df_final['rhs']+str(", ")+df_final['rhs2']+str(", ")+df_final['rhs3']
 df_final.drop(columns = ['lhs2','rhs2','rhs3'],inplace = True)
 df_final
```
