```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importing the dataset
data = pd.read_csv("/content/drive/MyDrive/4th SEM/ML/Datasets/CrabAgePrediction.csv")
data.head()
```

```python
#Checking missing values
data.isna().sum()
```

```python
#Splitting target and features
x = data.iloc[:,:-1].values
y = data.iloc[:,[-1]].values
```

```python
#Encoding categorical variables
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['Sex'] = le.fit_transform(data['Sex'])
```

```python
#Splitting train and test data
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
#Fitting the model
from sklearn.svm import SVR
rg = SVR(kernel="linear")
rg.fit(xtrain, ytrain)
```

```python
ypred = rg.predict(xtest)
```

```python
#Evaluation metrics
from sklearn.metrics import r2_score
print("R2 SCORE : ",r2_score(ytest, ypred))
```

SVM Classification

```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importing the dataset
data = pd.read_csv("/content/drive/MyDrive/4th SEM/ML/Datasets/heart_statlog_cleveland_hungary_final.csv")
data.head()
```

```python
#Checking missing values
data.isna().sum()
```

```python
#Splitting target and features
x = data.iloc[:,:-1].values
y = data.iloc[:, [-1]].values
```

```python
#Splitting train and test data
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x,y,test_size=0.2,random_state=3)
```

```python
#Fitting the model
from sklearn.svm import SVC
clf = SVC(kernel="linear")
clf.fit(xtrain,ytrain)
```

```python
y_pred = clf.predict(xtest)
```

```python
#Classification Report
from sklearn.metrics import classification_report
print(classification_report(ytest,y_pred))
```

```python
#Evaluation metrics
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
print("Accuracy:", accuracy_score(ytest, y_pred))
print("Precision:", precision_score(ytest, y_pred))
print("Recall Score:", recall_score(ytest, y_pred))
print("F1 Score:", f1_score(ytest, y_pred))
```
