```python
#Importing necessary libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Loading the dataset
data=pd.read_csv("/content/drive/MyDrive/4th SEM/Datasets/insurance_data - insurance_data.csv")
data.head()
```

```python
#Checking missing values
data.isna().sum()
```

```python
#Splitting data into feature and target
x=data[['age']]
y=data['bought_insurance']
```

```python
#Splitting data into train and test data
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state=7)
```

```python
#Plotting
plt.scatter(x,y)
plt.title("Plotting \nHari-717822i216")
plt.xlabel("Age")
plt.ylabel("Bought_Insurance")
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lg=LogisticRegression()
lg.fit(xtrain, ytrain)
```

```python
#Performance metrics of training set
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
ypred1 = lg.predict(xtrain)
print("Accuracy:", accuracy_score(ytrain, ypred1))
print("Precision:", precision_score(ytrain, ypred1))
print("Recall Score:", recall_score(ytrain,ypred1))
print("F1 Score:", f1_score(ytrain, ypred1))
```

```python
#Confusion matrix - order based on classes of target
from sklearn.metrics import confusion_matrix as cm
cm(ytrain, ypred1)
lg.predict_proba(xtrain)
```

```python
#Heatmap from Confusion matrix of training set
import seaborn as sns
sns.heatmap(cm(ytrain, ypred1), annot = True)
```

```python
#Classification Report for training data
from sklearn.metrics import classification_report as cr
print(cr(ytrain, ypred1))
```

```python
#Performance metrics of testing set
ypred2 = lg.predict(xtest)
print("Accuracy:", accuracy_score(ytest, ypred2))
print("Precision:", precision_score(ytest, ypred2))
print("Recall Score:", recall_score(ytest,ypred2))
print("F1 Score:", f1_score(ypred2, ytest))
```

```python
#Confusion matrix for testing set- order based on classes of target
from sklearn.metrics import confusion_matrix as cm
cm(ytest, ypred2)
```

```python
#Heatmap from Confusion matrix of training set
import seaborn as sns
sns.heatmap(cm(ytest, ypred2), annot = True)
```

```python
#Classification Report for testing data
from sklearn.metrics import classification_report as cr
print(cr(ytest, ypred2))
```
