```python
#Importing necessary modules
import numpy as np
from sklearn.datasets import load_digits
digits=load_digits()
```

```python
#Splitting target and features
X = digits.data
y = digits.target
```

```python
X.shape
```

```python
#Splitting train and test sets
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain,ytest = train_test_split(X, y, test_size=0.2, random_state=1)
```

```python
#Model fitting - Logistic Regression
from sklearn.linear_model import LogisticRegression
rg = LogisticRegression()
rg.fit(xtrain, ytrain)
```

```python
ypred1 = rg.predict(xtest)
```

```python
#Metrics
from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(ypred1, ytest))
```

```python
#LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda = LinearDiscriminantAnalysis(n_components=6)
xtrain_lda= lda.fit_transform(xtrain, ytrain)
xtest_lda = lda.transform(xtest)
```

```python
xtrain_lda.shape
```

```python
rg.fit(xtrain_lda, ytrain)
```

```python
ypred2 = rg.predict(xtest_lda)
```

```python
#Metrics
from sklearn.metrics import accuracy_score
print("Accuracy:", accuracy_score(ypred2, ytest))
```
