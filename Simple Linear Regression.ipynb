```python
def gradient_descent(x,y):
  m_curr=b_curr=0
  iterations=1000
  learning_rate=0.08
  n=len(x)
  y_p=[]

  for i in range(iterations):
    y_predicted=b_curr+m_curr*x
    y_p.append(y_predicted)
    cost=(1/n)*sum([val**2 for val in (y-y_predicted)])
    md=-(2/n)*sum(x*(y-y_predicted))
    bd=-(2/n)*sum(y-y_predicted)
    b_curr=b_curr-learning_rate*bd
    m_curr=m_curr-learning_rate*md
    print('m {}, b {}, cost {}, iteration{}'.format(m_curr,b_curr,cost,i))

  return y_p
#y=2x+3
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])

y_p=gradient_descent(x,y)
```

```python
plt.scatter(x,y)
for i in y_p:
  plt.plot(x,i)
plt.xlabel('x data')
plt.ylabel('y data')
plt.title('717822i216 Hari')
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
d=pd.read_csv('/content/drive/MyDrive/4th SEM/Datasets/salary - salary.csv')
```

```python
d.head()
```

```python
d.shape
```

```python
d.columns
```

```python
#Checking null values
d.isna().sum()
```

```python
#Handling null values
d['Salary'] = d['Salary'].fillna(d['Salary'].mean())
```

```python
#Splitting features and targets
x = d.iloc[:,:-1]
y = d.iloc[:,-1]
```

```python
#Splitting train and test data
from sklearn.model_selection import train_test_split
xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 3)
```

```python
#To check the correlation
d.corr()
```

```python
#Training the model
from sklearn.linear_model import LinearRegression
rg = LinearRegression()
rg.fit(xtrain, ytrain)
```

```python
#To identify m & c values
y_pred1 = rg.predict(xtest)
y_pred2 = rg.predict(xtrain)
```

```python
#Scatter plot for test data
plt.scatter(xtest, ytest)
plt.plot(xtest, y_pred1)
plt.xlabel('YearsOfExperience_xtest')
plt.ylabel('Salary_ytest')
plt.title('Testing Plot')
```

```python
#Scatter plot for train data
plt.scatter(xtrain, ytrain)
plt.plot(xtrain, y_pred2)
plt.xlabel('YearsOfExperience_xtrain')
plt.ylabel('Salary_ytrain')
plt.title('Training Plot')
```

```python
#To find the level of accuracy
rg.score(xtrain, ytrain), rg.score(xtest, ytest)
```
