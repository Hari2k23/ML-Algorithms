```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("/content/drive/MyDrive/4th SEM/Datasets/Position_Salaries - Position_Salaries.csv")
```

```python
data.head()
```

```python
data.isna().sum()
```

```python
x=data.iloc[:,1:2]
y=data.iloc[:,2:3]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size = 0.2,random_state = 1)
```

```python
from sklearn.linear_model import LinearRegression
rg = LinearRegression()
rg.fit(x_train, y_train)
```

```python
y_pred1=rg.predict(x_train)
plt.scatter(x_train, y_train)
plt.plot(x_train, y_pred1, color="magenta")
plt.xlabel("Level")
plt.ylabel("Salary")
plt.title("Linear Regression on train data")
```

```python
from sklearn.preprocessing import PolynomialFeatures
Pf = PolynomialFeatures(degree = 3)
xpoly = Pf.fit_transform(x)
```

```python
rg.fit(xpoly, y)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
print(mean_squared_error(y,rg.predict(xpoly)))
r2_score(y,rg.predict(xpoly))
```

```python
y_pred2=rg.predict(xpoly)
plt.scatter(x,y)
plt.plot(x, y_pred2)
plt.xlabel("Level")
plt.ylabel("Salary")
plt.title("Simple Polynomial Regression\n 717822I216 Hari")
```
